### (9) Use Boxplots or violinplots to visualise the distribution of change in life expectancy from 1952 to 2007.    

Hint tools from lab 6 may be helpful.

```python
#Created a dataframe to life expectancy for only year 1952
df_1952 = gapminder[gapminder['year']==1952]
```

```python
#Created a dataframe to life expectancy for only year 2007
df_2007 = gapminder[gapminder['year']==2007]
```

```python
#Used contact() method from pandas to combine these dataframes
Change_lifeExp = pd.concat([df_1952, df_2007], axis=0)
```

```python
Change_lifeExp
```

```python
#To show change in life Expectancy of two years, I have used the concatinated dafatframe and plotted its boxplot

ax = sns.boxplot(x='year', y='lifeExp', data=Change_lifeExp)
ax.set_title('Change in life expectancy from 1952 to 2007')
ax.set_xlabel('Year')
ax.set_ylabel('Life Expectancy')
```

### (10) Continue this analysis of the data. Include code cells, their outputs and markdown cells where you write about your findings. What patterns can you observe in the data?


```python
#Code cell

df_1952 = gapminder[gapminder['year']==1952]

df_2007 = gapminder[gapminder['year']==2007]

Change_lifeExp = pd.concat([df_1952, df_2007], axis=0)

ax = sns.boxplot(x='year', y='lifeExp', data=Change_lifeExp)

ax.set_title('Change in life expectancy from 1952 to 2007')

ax.set_xlabel('Year')

ax.set_ylabel('Life Expectancy')

```

## From above boxplot, we can see that, there is a substantial increase in the life Expectancy in year 2007 compared to year 1952; where the maximum life expectancy was 60 years in 1952, in 2007 it is somewhere near 77-78 years. The median value of life expectancy for year 1952 is around 55 years, whereas for year 2007 it is increased to somewhere around 73 years.

## Majority of the population had life expectancy between 55 to 60 years as per the upper quartile of year 1952 

## Majority of population had life expectancy between 58 to 73 years as per the lower quartile of year 2007

### There are no outliers for the data presented above
